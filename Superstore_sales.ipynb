# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv('datasets/Superstore.csv', encoding = 'unicode_escape')
```

# Display basic info

```python
print("Dataset Info:")
print(df.info())
```

# Sales and Profit by Category
# Insight: Helps identify which product categories are most profitable and deserve more investment.


```python
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
sns.barplot(x='Category', y='Sales', data=category_summary)
sns.barplot(x='Category', y='Profit', data=category_summary)
plt.show()
```

# Sales and Profit by Sub-Category
# Insight: Reveals top-performing sub-categories for targeted promotions.

```python
subcat_summary = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False).reset_index()
sns.barplot(x='Sub-Category', y='Sales', data=subcat_summary)
sns.barplot(x='Sub-Category', y='Profit', data=subcat_summary)
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Profit Heatmap by Region and State
# Insight: Shows geographic profitability, guiding regional strategy..

```python
region_state_profit = df.groupby(['Region', 'State'])['Profit'].sum().unstack().fillna(0)
sns.heatmap(region_state_profit, cmap='coolwarm')
plt.show()
```
